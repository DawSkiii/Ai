```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import tensorflow_datasets as tfds
import numpy as np
```

```python
ds_train, ds_test = tfds.load(
    "imdb_reviews",
    split=["train", "test"],
    as_supervised=True
)
```

```python
train_sentences = []
train_labels = []
test_sentences = []
test_labels = []

for sentence, label in tfds.as_numpy(ds_train):
    train_sentences.append(sentence.decode("utf-8"))
    train_labels.append(label)

for sentence, label in tfds.as_numpy(ds_test):
    test_sentences.append(sentence.decode("utf-8"))
    test_labels.append(label)

train_labels = np.array(train_labels)
test_labels = np.array(test_labels)
```

```python
bow_vectorizer = CountVectorizer(max_features=10000)
X_train_bow = bow_vectorizer.fit_transform(train_sentences).toarray()
X_test_bow = bow_vectorizer.transform(test_sentences).toarray()
```

```python
def build_model(input_dim):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model
```

```python
model_bow = build_model(X_train_bow.shape[1])
model_bow.fit(X_train_bow, train_labels,
              epochs=5, batch_size=512,
              validation_split=0.2)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
X_train_tfidf = tfidf_vectorizer.fit_transform(train_sentences).toarray()
X_test_tfidf = tfidf_vectorizer.transform(test_sentences).toarray()
```

```python
model_tfidf = build_model(X_train_tfidf.shape[1])
model_tfidf.fit(X_train_tfidf, train_labels,
                epochs=5, batch_size=512,
                validation_split=0.2)
```
